```python
import ast
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from pathlib import Path
from sklearn.preprocessing import MultiLabelBinarizer

plt.rcParams.update({
    "font.family": "DejaVu Sans",
    "axes.spines.top": False,
    "axes.spines.right": False,
    "axes.grid": True,
    "grid.alpha": 0.25,
    "grid.linestyle": "--"
})

print("✅ All libraries imported successfully.")
```

## 2. Load & Preprocess Test Results

```python
BASE = Path(r"c:\Users\eggoni\Desktop\llm")

MODEL_FILES = {
    "LLaMA 3.3 70b": BASE / "test_results.csv",
    "Gemma3 27b":    BASE / "test_results_gemma3_27b.csv",
    "LLaMA 3.1 8b":  BASE / "test_results_llama31_8b.csv",   # update filename if needed
}

def parse_list(raw):
    """Safely parse a stringified list into a Python list of strings."""
    if not isinstance(raw, str) or not raw.strip():
        return []
    try:
        val = ast.literal_eval(raw)
        if isinstance(val, (list, tuple)):
            return [str(x).strip() for x in val if str(x).strip()]
    except (ValueError, SyntaxError):
        pass
    return []

raw_dfs = {}
for model_name, csv_path in MODEL_FILES.items():
    if csv_path.exists():
        df = pd.read_csv(csv_path)
        df["pred"] = df["predicted_categories"].apply(parse_list)
        df["gt"]   = df["ground_truth"].apply(parse_list)
        raw_dfs[model_name] = df
        print(f"✅ Loaded {model_name}: {len(df)} rows from {csv_path.name}")
    else:
        print(f"⚠️  {csv_path.name} not found — skipping {model_name}")
```

## 3. Define Evaluation Metrics Helper Functions

```python
def calculate_metrics(preds, gts):
    """
    Calculate sample-level and micro/macro metrics for multi-label classification.
    preds, gts: lists of lists of strings
    """
    all_labels = sorted(set(lbl for lst in gts + preds for lbl in lst))
    mlb = MultiLabelBinarizer(classes=all_labels)
    mlb.fit([all_labels])

    Y_pred = mlb.transform(preds)
    Y_true = mlb.transform(gts)

    # Sample-level
    jaccard, precision, recall, f1 = [], [], [], []
    exact_match = 0
    hamming = []

    for yp, yt in zip(Y_pred, Y_true):
        inter = np.logical_and(yp, yt).sum()
        union = np.logical_or(yp, yt).sum()
        pred_sum = yp.sum()
        true_sum = yt.sum()

        jaccard.append(inter / union if union else 0)
        precision.append(inter / pred_sum if pred_sum else 0)
        recall.append(inter / true_sum if true_sum else 0)
        p = precision[-1]; r = recall[-1]
        f1.append(2 * p * r / (p + r) if (p + r) else 0)
        if np.array_equal(yp, yt):
            exact_match += 1
        hamming.append(np.logical_xor(yp, yt).mean())

    n = len(preds)

    # Micro F1
    tp_total = np.logical_and(Y_pred, Y_true).sum()
    fp_total = np.logical_and(Y_pred, ~Y_true.astype(bool)).sum()
    fn_total = np.logical_and(~Y_pred.astype(bool), Y_true).sum()
    micro_p = tp_total / (tp_total + fp_total) if (tp_total + fp_total) else 0
    micro_r = tp_total / (tp_total + fn_total) if (tp_total + fn_total) else 0
    micro_f1 = 2 * micro_p * micro_r / (micro_p + micro_r) if (micro_p + micro_r) else 0

    # Macro F1
    macro_f1_scores = []
    for i in range(Y_true.shape[1]):
        tp = np.logical_and(Y_pred[:, i], Y_true[:, i]).sum()
        fp = np.logical_and(Y_pred[:, i], ~Y_true[:, i].astype(bool)).sum()
        fn = np.logical_and(~Y_pred[:, i].astype(bool), Y_true[:, i]).sum()
        mp = tp / (tp + fp) if (tp + fp) else 0
        mr = tp / (tp + fn) if (tp + fn) else 0
        macro_f1_scores.append(2 * mp * mr / (mp + mr) if (mp + mr) else 0)

    return {
        "jaccard_similarity": np.mean(jaccard),
        "precision":          np.mean(precision),
        "recall":             np.mean(recall),
        "f1_score":           np.mean(f1),
        "exact_match_ratio":  exact_match / n,
        "hamming_loss":       np.mean(hamming),
        "micro_f1":           micro_f1,
        "macro_f1":           np.mean(macro_f1_scores),
    }


def per_label_report(preds, gts):
    """Return per-label precision, recall, F1, and support."""
    all_labels = sorted(set(lbl for lst in gts + preds for lbl in lst))
    mlb = MultiLabelBinarizer(classes=all_labels)
    mlb.fit([all_labels])
    Y_pred = mlb.transform(preds)
    Y_true = mlb.transform(gts)

    report = {}
    for i, label in enumerate(all_labels):
        tp = np.logical_and(Y_pred[:, i], Y_true[:, i]).sum()
        fp = np.logical_and(Y_pred[:, i], ~Y_true[:, i].astype(bool)).sum()
        fn = np.logical_and(~Y_pred[:, i].astype(bool), Y_true[:, i]).sum()
        p  = tp / (tp + fp) if (tp + fp) else 0
        r  = tp / (tp + fn) if (tp + fn) else 0
        f  = 2 * p * r / (p + r) if (p + r) else 0
        report[label] = {"precision": p, "recall": r, "f1": f, "support": int(Y_true[:, i].sum())}
    return report

print("✅ Metric helper functions defined.")
```

## 4. Evaluate Each Model

```python
def evaluate_file(df, model_name):
    mask     = (df["success"] == True) & (df["gt"].apply(len) > 0) & (df["pred"].apply(len) > 0)
    df_eval  = df[mask]
    failed   = len(df) - len(df_eval)
    preds    = df_eval["pred"].tolist()
    gts      = df_eval["gt"].tolist()
    metrics  = calculate_metrics(preds, gts)
    report   = per_label_report(preds, gts)
    return {
        "model":           model_name,
        "total":           len(df),
        "evaluated":       len(df_eval),
        "failed":          failed,
        "metrics":         metrics,
        "report":          report,
        "avg_pred_labels": sum(len(p) for p in preds) / len(preds),
        "avg_gt_labels":   sum(len(g) for g in gts)   / len(gts),
    }

results = {}
for model_name, df in raw_dfs.items():
    results[model_name] = evaluate_file(df, model_name)
    r = results[model_name]
    print(f"✅ {model_name}: evaluated={r['evaluated']}, failed={r['failed']}, "
          f"Micro F1={r['metrics']['micro_f1']:.4f}, Macro F1={r['metrics']['macro_f1']:.4f}")
```

## 5. Overall Metrics Comparison Table

```python
METRIC_MAP = [
    ("Precision",    "precision"),
    ("Recall",       "recall"),
    ("F1 (Micro)",   "micro_f1"),
    ("F1 (Macro)",   "macro_f1"),
    ("Jaccard",      "jaccard_similarity"),
    ("Hamming Loss", "hamming_loss"),
    ("Exact Match",  "exact_match_ratio"),
]

model_names = list(results.keys())

# Build overall DataFrame
overall_data = {"Metric": [d for d, _ in METRIC_MAP]}
for name in model_names:
    overall_data[name] = [results[name]["metrics"].get(k, 0) for _, k in METRIC_MAP]

overall = pd.DataFrame(overall_data)

# Pretty print
print(f"\n{'='*90}")
print(f"  MODEL COMPARISON — Overall Metrics")
print(f"{'='*90}")
header = f"  {'Metric':<22}"
for name in model_names:
    header += f"  {name:>15}"
header += f"  {'Winner':>16}"
print(header)
print(f"  {'-'*85}")

for i, (display, key) in enumerate(METRIC_MAP):
    row_str = f"  {display:<22}"
    vals = {name: results[name]["metrics"].get(key, 0) for name in model_names}
    for name in model_names:
        row_str += f"  {vals[name]:>15.4f}"
    winner = min(vals, key=vals.get) if key == "hamming_loss" else max(vals, key=vals.get)
    row_str += f"  ← {winner}"
    print(row_str)

print(f"{'='*90}")
overall
```
